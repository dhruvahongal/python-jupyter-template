### Exercise 5: Subplot Layout

```python
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.

import matplotlib.pyplot as plt
import numpy as np

# Create a figure and a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Data for the plots
x = np.linspace(0, 10, 100)
y_line = np.sin(x)  # Line chart data
categories = ['A', 'B', 'C', 'D']  # Bar chart categories
values = [5, 7, 3, 9]  # Bar chart values
x_scatter = [1, 2, 3, 4, 5]  # Scatter plot x values
y_scatter = [2, 4, 6, 8, 10]  # Scatter plot y values
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']  # Pie chart labels
pie_sizes = [40, 25, 20, 15]  # Pie chart sizes

# 1. Line chart in the first subplot
axs[0, 0].plot(x, y_line, color='blue')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')

# 2. Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# 3. Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter, color='purple')
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0,5), ha='center')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')

# 4. Pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Programming Language Popularity')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```
