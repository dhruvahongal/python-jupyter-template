### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import matplotlib.pyplot as plt

# Define the data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y, color='blue')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,5), ha='center')

# Set the title
plt.title('Scatter Plot with Annotations')

# Add labels for axes
plt.xlabel('X values')
plt.ylabel('Y values')

# Display the plot
plt.show()
```
